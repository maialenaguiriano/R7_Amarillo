```python
# Podemos especificar CSS personalizado para la App.
external_stylesheets = [
    "https://codepen.io/chriddyp/pen/bWLwgP.css"
]
path1 = '..\\datosTransformados\\' # ruta donde se van a guardar los datos limpios
df_limpio = pd.read_csv(path1 + 'datos_limpios2.csv', encoding = 'cp1252', index_col = [0])

# A la hora de introducir texto en la app, HTML es bastante tedioso. 
# Mucho más sencillo y rápido Markdown.

markdown_text_1 = """En esta pestaña se realizará un análisis de las empresas por sus valores en algunas de las variables. Con el fin de 
realizar una comparación más visual, abajo dispone de filtros que puede ajustar a su gusto."""

markdown_text_2 = """En esta pestaña se realizará un análisis por sectores. Con el fin de realizar una comparación más visual, 
abajo dispone de filtros que puede ajustar a su gusto."""

markdown_text_3 = """En esta pestaña se realizará un análisis por ratios. Con el fin de realizar una comparación más visual, 
abajo dispone de filtros que puede ajustar a su gusto."""

markdown_text_4 = """En esta pestaña se realizará un análisis de los resultados obtenidos de las predicciones realizadas. Con el fin de realizar 
una comparación más visual, abajo dispone de filtros que puede ajustar a su gusto."""

tabs_styles = {
    'backgroundColor':'#009aa6',
    'font-family': 'Times New Roman',
    'color': 'white'
}

# Creamos la aplicacion Dash y especificamos CSS.
app = dash.JupyterDash(
    __name__, 
    external_stylesheets = external_stylesheets
)

image_filename = 'elkargi-logo-header.png' 
encoded_image = base64.b64encode(open(image_filename, 'rb').read())

app.layout = html.Div(
    children = [
        html.Div(
            className = "row",
            children = [
                html.Img(
                    src='data:image/png;base64,{}'.format(encoded_image.decode()),
                    style = {
                        'display': 'block',
                        'margin-left': 'auto',
                        'margin-right': 'auto'
                    }
                    ),
                html.H4(
                    children = "Análisis interactivo del trabajo realizado por el grupo Amarillo",
                    style = {
                        "textAlign": "center"
                    }
                )
            ],
        ),
        dcc.Tabs([
            dcc.Tab( style = {'backgroundColor':'#ffb703', 'color':'black', 'borderTop': '10px solid #023047', 'font-family':'Stencil Std, fantasy', 'font-size': '25px'}, 
                selected_style= {'backgroundColor':'#023047', 'color':'white', 'borderTop': '10px solid #ffb703', 'font-family':'Stencil Std, fantasy', 'font-size': '25px'},
                label = "Descriptivos por tamaño",
                children = [
                    html.Div(
                        className = "row",
                        children = [
                            html.H5("Descripcion de la pestaña:"),
                            dcc.Markdown(children = markdown_text_1),
                            dcc.Markdown(id = "markdown-ebitda"),
                            dcc.Markdown(id = "markdown-activos")
                        ]
                    ),
                    html.Div(
                        className = "row",
                        children = [
                            html.Div(
                                children = [
                                    html.H6("Rango Ebitda"),
                                    dcc.RangeSlider(
                                        id = "range-ebitda",
                                        min = df_limpio["ebitda_mil_eur"].min(),
                                        max = df_limpio["ebitda_mil_eur"].max(),
                                        step = 2.5e4,
                                        value = [df_limpio["ebitda_mil_eur"].min(), df_limpio["ebitda_mil_eur"].max()]
                                    ),
                                    html.H6("Rango activos"),
                                    dcc.RangeSlider(
                                        id = "range-activo",
                                        min = df_limpio["total_activo_mil_eur"].min(),
                                        max = df_limpio["total_activo_mil_eur"].max(),
                                        step = 2.5e4,
                                        value = [df_limpio["total_activo_mil_eur"].min(), df_limpio["total_activo_mil_eur"].max()]
                                    )
                                ]
                            ),
                            html.Div(
                                children = [
                                    dcc.Graph(id = "sec-1", style = tabs_styles)
                                ],
                                className = "five columns"
                            ),
                            html.Div(
                                children = [
                                    dcc.Graph(id = "sec-2")
                                ],
                                className = "five columns"
                            ),
                            html.Div(
                                children = [
                                    dcc.Graph(id = "sec-3")
                                ],
                                className = "five columns"
                            )
                        ]
                    )
                ]
            ),
            dcc.Tab(style = {'backgroundColor':'#ffb703', 'color':'black', 'borderTop': '10px solid #023047', 'font-family':'Stencil Std, fantasy', 'font-size': '25px'}, 
                selected_style= {'backgroundColor':'#023047', 'color':'white', 'borderTop': '10px solid #ffb703', 'font-family':'Stencil Std, fantasy', 'font-size': '25px'},
                label = "Descriptivo por sector",
                children = [
                    html.Div(
                        className = "row",
                        children = [
                            html.H5("Descripcion de la pestaña:"),
                            dcc.Markdown(children = markdown_text_2)
                        ]
                    ),
                    html.Div(
                        className = "row",
                        children = [
                            html.Div(
                                children = [
                                    html.H6("Sector"),
                                    dcc.Dropdown(
                                        id = "dropdown-sector",
                                        options = [{
                                            "label": str(name),
                                            "value": name
                                        } for name  in list(np.sort(df_limpio["sector"].unique()))],
                                        value = [],
                                        placeholder = "Selecciona el sector:",
                                        multi = True
                                    ),
                                    html.H6("Estado"),
                                    dcc.Dropdown(
                                        id = "dropdown-estado",
                                        options = [{
                                            "label": str(name),
                                            "value": name
                                        } for name  in list(np.sort(df_limpio["estado"].unique()))],
                                        value = [],
                                        placeholder = "Selecciona el estado de la empresa:",
                                        multi = True
                                    ),
                                ]
                            ),
                            html.Div(
                                children = [
                                    dcc.Graph(id = "2-1", style = tabs_styles)
                                ]
                            )
                        ]
                    )]
            ),
            dcc.Tab(style = {'backgroundColor':'#ffb703', 'color':'black', 'borderTop': '10px solid #023047', 'font-family':'Stencil Std, fantasy', 'font-size': '25px'}, 
                selected_style= {'backgroundColor':'#023047', 'color':'white', 'borderTop': '10px solid #ffb703', 'font-family':'Stencil Std, fantasy', 'font-size': '25px'},
                label = "Polar",
                children = [
                    html.Div(
                        className = "row",
                        children = [
                            html.H5("Descripcion de la pestaña:"),
                            dcc.Markdown(children = markdown_text_3)
                        ]
                    ),
                    html.Div(
                        className = "row",
                        children = [
                            html.Div(
                                children = [
                                    html.H6("Sector"),
                                    dcc.Dropdown(
                                        id = "dropdown-sector2",
                                        options = [{
                                            "label": str(name),
                                            "value": name
                                        } for name  in list(np.sort(df_limpio["sector"].unique()))],
                                        value = [],
                                        placeholder = "Selecciona el sector:",
                                        multi = True
                                    ),
                                ]
                            ),
                            html.Div(
                                children = [
                                    dcc.Graph(id = "3-1", style = tabs_styles)
                                ]
                            )
                        ]
                    )]
            ),
            dcc.Tab(style = {'backgroundColor':'#ffb703', 'color':'black', 'borderTop': '10px solid #023047', 'font-family':'Stencil Std, fantasy', 'font-size': '25px'}, 
                selected_style= {'backgroundColor':'#023047', 'color':'white', 'borderTop': '10px solid #ffb703', 'font-family':'Stencil Std, fantasy', 'font-size': '25px'},
                label = "Conclusiones",
                children = [
                    html.Div(
                        className = "row",
                        children = [
                            html.H5("Descripcion de la pestaña:"),
                            dcc.Markdown(children = markdown_text_4)
                        ]
                    ),
                    html.Div(
                        className = "row",
                        children = [
                            html.Div(
                                children = [
                                    html.H6("Sector"),
                                    dcc.Dropdown(
                                        id = "dropdown-sector3",
                                        options = [{
                                            "label": str(name),
                                            "value": name
                                        } for name  in list(np.sort(df_limpio["sector"].unique()))],
                                        value = [],
                                        placeholder = "Selecciona el sector:",
                                        multi = True
                                    ),
                                ]
                            ),
                            html.Div(
                                children = [
                                    dcc.Graph(id = "4-1", style = tabs_styles)
                                ],
                                className = "five columns"
                            ),
                            html.Div(
                                children = [
                                    dcc.Graph(id = "4-2", style = tabs_styles)
                                ],
                                className = "five columns"
                            )
                        ]
                    )]
            ),
        ], style=tabs_styles)
    ]
, style = tabs_styles)

'''
PRIMERA PESTAÑA
'''
@app.callback(
    Output(
        component_id = "sec-1", 
        component_property = "figure"
    ),
    Output(
        component_id = "sec-2", 
        component_property = "figure"
    ),
    Output(
        component_id = "sec-3", 
        component_property = "figure"
    ),
    Output(
        component_id = "markdown-ebitda",
        component_property = "children"
    ),
    Output(
        component_id = "markdown-activos",
        component_property = "children"
    ),
    Input(
        component_id = "range-ebitda",
        component_property = "value"
    ),
    Input(
        component_id = "range-activo",
        component_property = "value"
    )
)

def update_figure(selected_ebitda, selected_activos):
    dff_limpio = df_limpio[df_limpio["ebitda_mil_eur"].between(*selected_ebitda)]
    dff_limpio = dff_limpio[dff_limpio["total_activo_mil_eur"].between(*selected_activos)]

    # Creamos la figura.
    fig1 = px.pie(
        dff_limpio, 
        names = 'sector', 
        hole = 0.5
    )

    # Actualizamos el layout.
    fig1.update_layout(
        title = "Porcentaje sectores"
    )

    # Figura
    fig2 = px.scatter(
        dff_limpio, 
        x = "ebitda_mil_eur", 
        y = "total_activo_mil_eur", 
        color = "sector"
    )

    # Layout 
    fig2.update_layout(
        title = "", 
        xaxis_title = "Ebitda", 
        yaxis_title = "Activos"
    )

    # Preparamos los datos.
    df_3 = df_limpio
    df_3['net_debt/ebitda'] = [ round(elem, 0) for elem in df_3['net_debt/ebitda'] ]
    df_3 = df_3[(df_3['net_debt/ebitda'] > 0) & (df_3['net_debt/ebitda'] < 15)]
    df_3 = df_3.groupby('sector')['net_debt/ebitda'].value_counts().unstack('net_debt/ebitda', fill_value=0).reset_index().iloc[:,1:]
    dff_3 = np.array(df_3)

    # Creamos la figura.
    fig3 = px.imshow(dff_3, aspect='auto', labels=dict(x="Net Debt/EBITDA", y="Sector", color="Nº de compañías"),
            y = ['Agricultura', 'Construccion', 'Deporte', 'Hosteleria',
       'Industria', 'Servicios', 'Transporte'])

    # Actualizamos el layout.
    fig3.update_layout(
        title = "",
        xaxis_title = "Deuta neta/Ebitda",
        yaxis_title = "Sector"
    )

    # Actualizamos el texto Markdown
    text_ebitda = """El **rango de EBITDA** está entre {0:.0f} y {1:.0f}.""".format(*selected_ebitda)
    text_activos = """El **rango de activos** está entre {0:.0f} y {1:.0f}.""".format(*selected_activos)

    return fig1, fig2, fig3, text_ebitda, text_activos

'''
SEGUNDA PESTAÑA
'''
@app.callback(
    Output(
        component_id = "2-1", 
        component_property = "figure"
    ),
    Input(
        component_id = "dropdown-sector",
        component_property = "value"
    ),
    Input(
        component_id = "dropdown-estado",
        component_property = "value"
    )
)
def update_figure(select_sector, select_estado):
    dff_limpio = df_limpio[df_limpio['sector'].isin(select_sector)]
    dff_limpio = dff_limpio[dff_limpio['estado'].isin(select_estado)]

    # Creamos la figura.
    fig4 = px.histogram(
        dff_limpio, 
        x = 'estado',
        color = 'sector'
    )

    # Actualizamos el layout.
    fig4.update_layout(
        title = "",
        xaxis_title = "Estado de la empresa",
        yaxis_title = "Número de empresas",
        bargap = 0.1
    ) 

    return fig4

'''
TERCERA PESTAÑA
'''
@app.callback(
    Output(
        component_id = "3-1", 
        component_property = "figure"
    ),
    Input(
        component_id = "dropdown-sector2",
        component_property = "value"
    )
)

def update_figure(select_sector):
    dff_limpio = df_limpio[df_limpio['sector'].isin(select_sector)]
    df_5 = dff_limpio[['ratio_liquidez', 'ratio_endeudamiento', 'ratio_tesoreria', 'net_debt/ebitda', 'ROE']]
    df_5.replace([np.inf, -np.inf], np.nan, inplace=True)
    df_5.loc[df_5['net_debt/ebitda'] < 0, 'net_debt/ebitda'] = np.nan
    df_5.loc[df_5['net_debt/ebitda'] > 20, 'net_debt/ebitda'] = np.nan
    df_5 = df_5.dropna()
    df_5.ROE = df_5.ROE * 100
    df_5 = pd.DataFrame(df_5.median(axis=0), columns=['valores'])
    df_5 = df_5.rename_axis('ratio').reset_index()
    
    fig5 = px.line_polar(df_5, line_close=True, theta='ratio', r= 'valores',
                        color_discrete_sequence=px.colors.sequential.Plasma_r,
                        template="plotly_dark")

    return fig5

'''
CUARTA PESTAÑA
'''
@app.callback(
    Output(
        component_id = "4-1", 
        component_property = "figure"
    ),
    Output(
        component_id = "4-2", 
        component_property = "figure"
    ),
    Input(
        component_id = "dropdown-sector3",
        component_property = "value"
    )
)

def update_figure(select_sector):
    dff_limpio = df_limpio[df_limpio['sector'].isin(select_sector)]
    
    fig6 = px.histogram(
        dff_limpio, 
        x = "quiebra", 
        color = "sector"
    )
    fig6.update_layout(
        title = "", 
        xaxis_title = "Quiebra", 
        yaxis_title = "Nº de empresas", 
        bargap = 0.1
    )

    df_6 = dff_limpio

    df_6.loc[df_6['ratio_endeudamiento'] < 0, 'ratio_endeudamiento'] = np.nan
    df_6.loc[df_6['ratio_endeudamiento'] > 5, 'ratio_endeudamiento'] = np.nan

    df_6.loc[df_6['ratio_liquidez'] < 0, 'ratio_liquidez'] = np.nan
    df_6.loc[df_6['ratio_liquidez'] > 7, 'ratio_liquidez'] = np.nan

    df_6.loc[df_6['net_debt/ebitda'] < 0, 'net_debt/ebitda'] = np.nan
    df_6.loc[df_6['net_debt/ebitda'] > 20, 'net_debt/ebitda'] = np.nan

    df_6 = df_6.dropna()

    fig7 = px.scatter(df_6, x="ratio_liquidez", y="net_debt/ebitda", size = 'ratio_endeudamiento',  color="quiebra")
    fig7

    return fig6, fig7

# Solamente se ejecuta si ejecutamos este script directamente.
if __name__ == "__main__":
    app.run_server(mode="external", debug=True, port = 1824)
```
